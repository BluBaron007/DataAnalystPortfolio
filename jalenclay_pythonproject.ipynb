```python
import seaborn as sns
sns.set_theme()

mpg = sns.load_dataset('mpg')

```

### 1. Relationship Between Horsepower & Fuel Economy

```python
sns.relplot(data=mpg, x="horsepower", y='mpg')
```

#### Summary: Horsepower is directly related to the mpg of the vehicle. The higher the HP the less fuel efficient it becomes. 

### 2. Fuel Economy By Country Of Origin

```python
mpg.groupby(by='origin').mpg.max().sort_values(ascending=False)
```

##### Japan has the most fuel efficient vehicles available

### 3. How Fuel Economy Has Changed Over Time

```python
sns.set_style('darkgrid')
sns.lineplot(data=mpg, x="model_year", y='mpg')
```

#### Summary: Vehicles have become more fuel efficient with time

### 4. Create Two Other Visuals Or Tables Realted To Fuel Economy

#### MPG vs Vehicle Weight

```python
sns.set_style('whitegrid')
palette = {"#FF5733"}
sns.set_palette(palette)
sns.scatterplot(data=mpg, x="weight", y='mpg')
```

#### Summary: MPG is higher when the vehicle weighs less than 4000lbs 

#### MPG vs Vehicle Displacement

```python
sns.set_style('whitegrid')
palette = {"#FF5745"}
sns.set_palette(palette)
sns.lineplot(data=mpg, x="mpg", y='displacement')
```

#### Summary: MPG is lower as the vehicle displacment increases in size. In other words, the most powerful engines are less fuel efficient but have the highest capabilities.
